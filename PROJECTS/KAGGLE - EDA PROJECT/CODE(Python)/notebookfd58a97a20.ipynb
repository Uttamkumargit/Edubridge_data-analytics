# Data visualization

```python
plt.figure(figsize=(15,10))
plot = sns.barplot(data['Liters of pure alcohol consumed per capita'], data['Country'],edgecolor='red')
for i,(value,name) in enumerate(zip(data['Liters of pure alcohol consumed per capita'],data['Country'])):
    plot.text(value,i-0.05,f'{value:,.0f}',size=10)
plt.title(' Alchocal Consumption by Country Wise ',fontsize=30,color='red', fontweight='bold')    
plt.show()
```

# to display most consumption acohal

```python
# pi#pie chart
x=data['Country'].head(10)
y=data['Liters of pure alcohol consumed per capita'].head(10)
plt.pie(y,labels=x,radius=1.2,autopct='%0.01f%%',shadow=True,explode=[.05,.2,.05,.2,.05,.05,.2,.05,.2,.05])
plt.show()
```

```python
px.bar(x='Country', y='Liters of pure alcohol consumed per capita', data_frame=data.head(15), title="Top 15 Countries with highest Alcohal Consumption")
```

```python
plt.figure(figsize = (10,6))
sns.histplot(data['Liters of pure alcohol consumed per capita'],color = 'blue', edgecolor = 'black')
plt.title("Distribution of Alcohal Consumption")
plt.show()
```
